```python
th = np.linspace(0,2*np.pi)
x , y = np.cos(th), np.sin(th)
x_esp, y_esp =th/(2*np.pi)*np.cos(th), th/(2*np.pi)*np.sin(th)
plt.plot(x,y, label=r'Crculo')
plt.plot(x_esp, y_esp,label=r'Espiral')
plt.gca().set_aspect('equal')
plt.show()
```

Campo electrico de una carga es
$$ 
\vec(E) = \frac{1}{4\pi\epsilon_0} \frac{q(\vec r - \vec r_q)}{|\vec r - \vec r_q|^2} 
$$

```python
def E_q(r,r_q,q):
    d32 = ((r[0]-r_q[0])**2+(r[1]-r_q[1])**2)**1.5
    #d32 es la norma al cuadrado?
    return (q*(r[0]-r_q[0]))/d32, (q*(r[1]-r_q[1]))/d32
```

```python
x,y = np.linspace(-2,2), np.linspace(-2,2)
XX, YY = np.meshgrid(x,y)
#este comando nos permite hacer las todas las combinaciones de parejas
N = np.sqrt(XX**2 + YY**2)
#Dividimos entre N para que todos los vectores sean unitarios
plt.quiver(XX, YY, YY/N, -XX/N)
plt.streamplot(XX, YY, YY/N, -XX/N)
#Nos muestra la lineas de campo
plt.gca().set_aspect('equal')
plt.show()
```

```python
r_q, q=[0,0], 1
Ex,Ey =E_q([XX,YY],r_q,q)
plt.streamplot(XX,YY,Ex,Ey, density=2)
plt.plot(r_q[0],r_q[1], 'o')
plt.gca().set_aspect('equal')
plt.show()
```

```python
r_q, q=[0,1], 1
Ex,Ey =E_q([XX,YY],r_q,q)
r_q_n, q_n=[0,-1], -1
Ex_neg,Ey_neg = E_q([XX,YY], r_q_n, q_n)
Et_x = Ex + Ex_neg
Et_y= Ey + Ey_neg
plt.streamplot(XX,YY,Et_x,Et_y, density=2)
plt.plot(r_q[0],r_q[1], 'o')
plt.gca().set_aspect('equal')
plt.show()
```
